## Variables and Data Types
**What are variables?**
Think of variables as labels for containers that store data. You assign a value to a variable using the equals sign `=`.

**Naming conventions:**
* Names should be descriptive (e.g., `sample_id` instead of `s`).
* Start with a letter or underscore (`_`).
* Can contain letters, numbers, and underscores.
* Case-sensitive (`sample_id` is different from `Sample_ID`).
* Avoid using Python keywords (e.g., `print`, `if`, `for`).

**Common Data Types:**
* **Integer (`int`):** Whole numbers, e.g., `42`, `-10`.
* **Float (`float`):** Numbers with a decimal point, e.g., `3.14`, `-0.001`, `2.0`.
* **String (`str`):** Text, enclosed in single quotes (`'...'`) or double quotes (`"..."`), e.g., `"DNA"`, `'Escherichia coli'`.
* **Boolean (`bool`):** Represents truth values, either `True` or `False`.

We can use the `type()` function to check the data type of a variable.




````python
# Exercise 1.1: Storing Scientific Data

# Create variables for a fictitious experiment
project_name = "Metabolic Response Study"
number_of_samples = 96
ph_level = 7.4
results_preliminary = True # Boolean: True or False

# Print the variables
print("Project Name:", project_name)
print("Number of Samples:", number_of_samples)
print("pH Level:", ph_level)
print("Are results preliminary?", results_preliminary)

# Print the data types of these variables
print("\nData type of project_name:", type(project_name))
print("Data type of number_of_samples:", type(number_of_samples))
print("Data type of ph_level:", type(ph_level))
print("Data type of results_preliminary:", type(results_preliminary))

# Instructor Note: Ask participants to create their own variables for a hypothetical experiment they might run.
````

### Basic Operators

Python supports various [operators](https://docs.python.org/3/library/operator.html) for performing calculations and comparisons.

**Arithmetic Operators:**
* `+` (addition)
* `-` (subtraction)
* `*` (multiplication)
* `/` (division - always results in a float)
* `//` (floor division - discards the fractional part)

* `==` (equal to)
* `!=` (not equal to)

* `<=` (less than or equal to)

**Logical Operators:** (Combine Boolean values)
* `and` (True if both are True)
* `or` (True if at least one is True)
* `not` (Inverts the Boolean value)

### Introduction to Data Structures: Lists

A **list** is an ordered, mutable (changeable) collection of items. One of the many [data structures](https://docs.python.org/3/tutorial/datastructures.html) found in Python (and other programming languages). Lists can contain items of different data types.

* **Creating lists:** `my_list = [item1, item2, item3]`
* **Accessing elements:** Use square brackets `[]` with an index. Python is 0-indexed, meaning the first item is at index `0`.
    * `my_list[0]` gives the first item.
    * `my_list[-1]` gives the last item.
* **Slicing:** Get a sub-list: `my_list[start_index:end_index]` (end_index is exclusive).
* **Modifying lists:**
    * `append(item)`: Adds an item to the end.
    * `insert(index, item)`: Inserts an item at a specific index.
    * `remove(item)`: Removes the first occurrence of an item.
    * `pop(index)`: Removes and returns the item at an index (default is the last item).
* **Length of a list:** `len(my_list)`

````python
# Exercise 1.3: Managing Experimental Readings

# List of temperature readings
temperature_readings_celsius = [22.5, 22.7, 22.8, 23.1, 22.9, 23.5, 100.2] # Oops, one looks wrong!
print("Initial readings:", temperature_readings_celsius)

# Access the first and last readings
first_reading = temperature_readings_celsius[0]
last_reading = temperature_readings_celsius[-1] # Python's way to get the last item
print("First reading:", first_reading)
print("Last reading (before correction):", last_reading)

# Add a new reading
temperature_readings_celsius.append(23.0)
print("After appending 23.0:", temperature_readings_celsius)

# Remove an erroneous reading (e.g., 100.2)
temperature_readings_celsius.remove(100.2) # This removes the first occurrence
# If you don't know the value but know the index (e.g. index 6 was wrong):
# temperature_readings_celsius.pop(6)
print("After removing 100.2:", temperature_readings_celsius)

# Calculate the number of readings taken
number_of_readings = len(temperature_readings_celsius)
print("Number of readings:", number_of_readings)

# Slicing: Get the first three readings
first_three_readings = temperature_readings_celsius[0:3] # index 3 is not included
print("First three readings:", first_three_readings)

# Instructor Note: Show how indexing beyond the list length causes an error.
# Explain 0-indexing carefully. Use analogies like floors in a building if helpful.
````

### Conditional Statements (If-Else)

Conditional statements allow your program to make decisions based on whether certain conditions are `True` or `False`.

**Syntax:**
```python
if condition1:
    # Code to execute if condition1 is True
elif condition2:  # Optional: "else if"
    # Code to execute if condition1 is False and condition2 is True
else:             # Optional
    # Code to execute if all preceding conditions are False
```
Indentation is crucial in Python! It defines blocks of code.

````python
# Exercise 2.1: Classifying Results

# Example 1: P-value classification
p_value = 0.065 # Try changing this value (e.g., 0.03, 0.15)

print(f"P-value is: {p_value}") # f-strings are a nice way to format strings!
if p_value < 0.05:
    print("Result is statistically significant.")
elif p_value < 0.1: # This means 0.05 <= p_value < 0.1
    print("Result is approaching significance (monitor closely).")
else:
    print("Result is not statistically significant.")

# Example 2: Cell culture confluence
confluence_percentage = 85 # Try 95, 60

print(f"\nCell confluence: {confluence_percentage}%")
if confluence_percentage > 90:
    print("Ready to split cells.")
elif confluence_percentage >= 70: # 70 to 90 (inclusive of 70, exclusive of >90 handled above)
    print("Monitor culture closely.")
else:
    print("Continue incubation.")

# Instructor Note: Emphasize the importance of indentation.
# Ask learners to trace the logic with different input values.
````

### Loops

Loops are used to repeat a block of code multiple times.

**`for` loops:** Iterate over a sequence (like a list, string, or a range of numbers).
```python
for variable_name in sequence:
    # Code to execute for each item in the sequence
```
The `range()` function is useful for generating sequences of numbers:
* `range(stop)`: 0 up to (but not including) `stop`. E.g., `range(5)` -> 0, 1, 2, 3, 4
* `range(start, stop)`: `start` up to (but not including) `stop`.
* `range(start, stop, step)`: `start` up to `stop`, incrementing by `step`.

**`while` loops:** Repeat code as long as a condition is `True`.
```python
while condition:
    # Code to execute
    # IMPORTANT: Ensure the condition eventually becomes False, or you'll have an infinite loop!
```

````python
# Exercise 2.2: Processing Data Collections

# Example 1: `for` loop with a list of gene names
gene_names = ["TP53", "BRCA1", "EGFR", "MYC"]

print("Analyzing genes:")
for gene in gene_names:
    print(f"  Processing gene: {gene}")

# Example 2: `for` loop with `range()` for a time-course experiment
print("\nSimulating time-course experiment:")
number_of_time_points = 5
for time_point in range(1, number_of_time_points + 1): # range(1, 6) will give 1, 2, 3, 4, 5
    print(f"Time point {time_point}: Taking measurement...")

# Example 3: `while` loop to simulate a reaction
# Let's say we want to reach a target concentration of at least 0.9
current_concentration = 0.5
target_concentration = 0.9
reaction_step = 0
print("\nSimulating reaction until target concentration is met:")
while current_concentration < target_concentration:
    reaction_step += 1 # Same as reaction_step = reaction_step + 1
    current_concentration += 0.08 # Simulate an increase in concentration
    # To avoid too many print statements if values are off, cap the steps for this demo
    if reaction_step > 100:
        print("Stopping: Took too many steps (safeguard).")
        break # Exit the loop immediately
    print(f"Step {reaction_step}: Current concentration = {current_concentration:.2f}") # .2f formats float to 2 decimal places

if current_concentration >= target_concentration:
    print(f"Target concentration of {target_concentration} reached or exceeded at step {reaction_step}.")
else:
    print(f"Target concentration not reached after {reaction_step} steps.")

# Instructor Note: Discuss infinite loops and how to break them in Colab (Interrupt execution button).
# For the `while` loop, highlight the part that changes the condition (`current_concentration += 0.08`).
````

### Functions

Functions are named, reusable blocks of code that perform a specific task. They help make your code more organized, readable, and efficient.

**Defining a function:**
```python
def function_name(parameter1, parameter2, ...):
    """
    Docstring: Explain what the function does, its parameters, and what it returns.
    This is a good practice!
    """
    # Code block for the function
    # ...
    return some_value  # Optional: to send a result back
```
* `def` keyword starts a function definition.
* `function_name`: Choose a descriptive name.
* `parameters` (optional): Input values the function works with.
* **Docstring:** A string literal right after the function header, used to document the function. Access with `help(function_name)`.
* `return` statement (optional): Sends a value back to the part of the code that called the function. If omitted, the function returns `None`.

**Calling a function:**
`result = function_name(argument1, argument2, ...)`
Arguments are the actual values passed to the function's parameters.

````python
# Code cell
# Exercise 2.3: Creating Reusable Scientific Tools

# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    """Converts temperature from Celsius to Fahrenheit.

    Args:
        celsius (float or int): Temperature in Celsius.

    Returns:
        float: Temperature in Fahrenheit.
    """
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

# Test the function
temp_c = 25.0
temp_f = celsius_to_fahrenheit(temp_c)
print(f"{temp_c}°C is equal to {temp_f}°F")

temp_c_boiling = 100.0
print(f"Water boils at {celsius_to_fahrenheit(temp_c_boiling)}°F at sea level.")


# Function to calculate the average of a list of numbers
def calculate_average(numbers_list):
    """Calculates the average of a list of numbers.

    Args:
        numbers_list (list of float or int): A list of numbers.

    Returns:
        float: The average of the numbers, or 0 if the list is empty.
    """
    if not numbers_list: # Check if the list is empty to avoid division by zero
        return 0.0
    total_sum = sum(numbers_list) # sum() is a built-in Python function
    count = len(numbers_list)
    average = total_sum / count
    return average

# Test the average function with our earlier temperature readings
# (Let's redefine it here for self-containment of the cell)
temperature_readings_celsius = [22.5, 22.7, 22.8, 23.1, 22.9, 23.5, 23.0]
avg_temp = calculate_average(temperature_readings_celsius)
print(f"\nAverage temperature: {avg_temp:.2f}°C")

empty_list = []
avg_empty = calculate_average(empty_list)
print(f"Average of an empty list: {avg_empty}")

# Show how to access the docstring
help(celsius_to_fahrenheit)
# print(celsius_to_fahrenheit.__doc__) # Another way

# Instructor Note: Explain local scope (variables inside a function are generally not accessible outside).
# Emphasize the DRY principle (Don't Repeat Yourself) and how functions help.
````

### Introduction to Libraries

Python's power comes from its extensive collection of **libraries** (also called modules or packages). These are collections of pre-written code that provide additional functionality.

**Importing libraries:**
* `import library_name`: Makes the whole library available. Access functions like `library_name.function()`.
* `import library_name as alias`: Imports the library and gives it a shorter alias (e.g., `import numpy as np`). This is very common.
* `from library_name import specific_function_or_item`: Imports only specific parts of a library. You can then use them directly, e.g., `specific_function()`.
* `from library_name import *`: Imports everything (generally discouraged for larger libraries as it can lead to name conflicts).

We will be using several libraries commonly used in scientific computing starting with NumPy and Pandas.

### NumPy for Numerical Data (Brief Introduction)

[**NumPy**](https://numpy.org/) is the fundamental package for numerical computation in Python. It provides:
* A powerful N-dimensional array object (the `ndarray`).
* Sophisticated (broadcasting) functions.
* Tools for integrating C/C++ and Fortran code.
* Useful linear algebra, Fourier transform, and random number capabilities.

The core of NumPy is the `ndarray`. NumPy arrays are more efficient for numerical operations than Python lists, especially for large datasets. They are also homogenous, meaning all elements must be of the same data type.

````python
# Code cell
# Exercise 3.1: Working with Numerical Arrays

# Our temperature readings from before
temperature_list_py = [22.5, 22.7, 22.8, 23.1, 22.9, 23.5, 23.0]

# Create a NumPy array from the Python list
temperature_array_np = np.array(temperature_list_py)
print("NumPy array of temperatures (Celsius):", temperature_array_np)
print("Type of the array:", type(temperature_array_np))
print("Data type of elements in the array:", temperature_array_np.dtype)

# Perform a simple calculation on the array (element-wise operation)
# Convert all Celsius readings to Kelvin (Kelvin = Celsius + 273.15)
kelvin_temperatures = temperature_array_np + 273.15
print("Temperatures in Kelvin:", kelvin_temperatures)

# NumPy makes math operations easy:
mean_temp_np = np.mean(temperature_array_np)
std_dev_temp_np = np.std(temperature_array_np)
max_temp_np = np.max(temperature_array_np)

print(f"\nUsing NumPy functions:")
print(f"Mean temperature: {mean_temp_np:.2f}°C")
print(f"Standard deviation: {std_dev_temp_np:.2f}°C")
print(f"Maximum temperature: {max_temp_np:.2f}°C")

# Compare with manual calculation or Python's sum()/len() for mean
# average_python = sum(temperature_list_py) / len(temperature_list_py)
# print(f"Mean temperature (Python list): {average_python:.2f}°C")

# Instructor Note: Briefly mention that operations on NumPy arrays are usually much faster than on Python lists for large datasets due to optimized C code underneath.
# Show creating an array with a specific dtype: np.array([1, 2, 3], dtype=float)
````

### Pandas for Tabular Data (Brief Introduction to Exploratory Data Analysis)

**Pandas** is a powerful, open-source library built on top of NumPy, providing fast, flexible, and expressive data structures designed to make working with "Panel Data" or "relational" or "labeled" data (like tables) both easy and intuitive. It is the go-to tool for Python-aligned scientits starting with EDA.

**Core Data Structures:**
* **Series:** A one-dimensional labeled array capable of holding any data type (integers, strings, floating point numbers, Python objects, etc.). Think of it as a single column in a table.
* **DataFrame:** A two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns). Think of it as a spreadsheet or SQL table.

**Common tasks with Pandas:**
* Reading and writing data from various file formats (CSV, Excel, SQL databases, etc.).
* Cleaning, filtering, and transforming data.
* Analyzing and summarizing data.
* Merging and joining datasets.
* Basic plotting
* Exploratory Data Analysis (EDA)

````python
# Exercise 3.2: Exploring a Small Dataset

# Create a dictionary with some sample experimental data
# Dictionaries are another Python data structure: key-value pairs.
data_dict = {
    'SampleID': ['S1', 'S2', 'S3', 'S4', 'S5'],
    'Treatment': ['Control', 'DrugA', 'Control', 'DrugB', 'DrugA'],
    'Measurement_OD': [0.15, 0.78, 0.20, 0.95, 0.65],
    'Replicate': [1, 1, 2, 1, 2]
}
print("This is a Python dictionary:")
print(data_dict)

# Convert this dictionary into a Pandas DataFrame
df_experiment = pd.DataFrame(data_dict)

print("\nThis is a Pandas DataFrame:")
print(df_experiment) # Colab often provides a nice HTML rendering for DataFrames

# Basic DataFrame inspection:
print("\nFirst 3 rows of the DataFrame (.head(3)):")
print(df_experiment.head(3))

print("\nLast 2 rows of the DataFrame (.tail(2)):")
print(df_experiment.tail(2))

print("\nInformation about the DataFrame (.info()):")
df_experiment.info() # Gives data types, non-null counts, memory usage

print("\nDescriptive statistics for numerical columns (.describe()):")
print(df_experiment.describe())

print("\nShape of the DataFrame (rows, columns) (.shape):")
print(df_experiment.shape)

# Accessing a single column (returns a Pandas Series)
treatment_column = df_experiment['Treatment']
print("\nThe 'Treatment' column (a Pandas Series):")
print(treatment_column)

print("\nUnique values in the 'Treatment' column:")
print(df_experiment['Treatment'].unique())

# Basic filtering: Samples treated with DrugA
drug_a_samples = df_experiment[df_experiment['Treatment'] == 'DrugA']
print("\nSamples treated with DrugA:")
print(drug_a_samples)


# Instructor Note: This is a very brief intro. The goal is to show what a DataFrame is and how to do basic manipulations.
# Suggestion for self-study:
# "You can load your own data from a CSV file using `pd.read_csv('your_file.csv')`."
# "If you have a CSV file on your computer, you can upload it to Colab using the file browser on the left panel, then read it."
# For example:
# from google.colab import files
# uploaded = files.upload() # This will open a dialog
# import io
# df_from_csv = pd.read_csv(io.BytesIO(uploaded['your_file_name.csv']))
# print(df_from_csv.head())
````
